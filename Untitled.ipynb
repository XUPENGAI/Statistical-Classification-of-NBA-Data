```python
import pandas as pd
import numpy as np
bball_file_name = 'nba.games.stats.csv'
ball = pd.read_csv(bball_file_name, sep = ',', decimal = '.')
```

```python
ball['Date'] = pd.to_datetime(ball['Date'])
ball.index = ball['Date']
del ball['Date']
ball.sort_index(inplace=True)
ball = ball[(ball['Home'] == "Home")]
ball = ball.drop('Unnamed: 0',axis=1)
ball.shape
print(ball)
null_data = ball[ball.isnull().any(axis=1)]
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
target = pd.DataFrame(ball['WINorLOSS'])
data = pd.DataFrame(ball[['OffRebounds', 'TotalRebounds', 'Assists', 'Steals', 'Blocks', 'Turnovers', 'TotalFouls', 'Opp.OffRebounds', 'Opp.TotalRebounds','Opp.Assists', 'Opp.Steals', 'Opp.Blocks', 'Opp.Turnovers', 'Opp.TotalFouls']])
data.shape

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3, random_state=0) 
classifier = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=5,
            max_features=None, max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, presort=False, random_state=None,
            splitter='best')  
classifier.fit(X_train, np.ravel(y_train,order='C'))
y_pred = classifier.predict(X_test) 
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred)) 
```

```python
from sklearn.utils import shuffle
import numpy as np
new_Ind = []
cur_MaxScore = 0.0
col_num = data.shape[1]
data2 = data.values
target2 = target.values
n_features=14
col_Ind_Random = shuffle(range(0,col_num), random_state=0)
for cur_f in range(0, col_num):
    new_Ind.append(col_Ind_Random[cur_f])
    newData = data2[:, new_Ind]
    X_train, X_test, y_train, y_test = train_test_split(newData, target2, test_size=0.3, random_state=0)
    clf1 = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=5,
            max_features=None, max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, presort=False, random_state=None,
            splitter='best')
    fit = clf1.fit(X_train, np.ravel(y_train,order='C'))
    cur_Score = clf1.score(X_test, y_test)
    if cur_Score < cur_MaxScore:
        new_Ind.remove(col_Ind_Random[cur_f])
    else:
        cur_MaxScore = cur_Score
    print "Decision Tree with ", cur_f +1, new_Ind, " selected features: " + str(cur_Score)
```
